```python
# Display the following menu to the user:
# 1) Addition
# 2) Subtraction
# Enter 1 or 2:
# If they enter a 1, it should run a subprogran that wil generate two random numbers between 5 and 20, and ask the user to add them together.
# Work out the correct answer and return both the user's answer and the correct answer.
# If they entered 2 as their selection on the menu, it should run a subprogram that will generate one number between 25 and 50 and another number
# between 1 and 25 and ask them to work out num1 minus num2. This way they will not have to worry about negative answer.
# Return both the user's answer and the correct answer. 
# Create another subprogram that will check if the user's answer matches the actual answer.
# If it does, display 'Correct', otherwise display a message that will say 'Incorrect, the answer is'
# and display the real answer. If they do not select a relevant option on the first menu you should display a suitable message.
import random

def addition():
    comp_num1 = random.randint(5,20)
    comp_num2 = random.randint(5,20)
    print(f'{comp_num1} + {comp_num2} = ')
    actual_answer = comp_num1 + comp_num2
    user_answer = int(input('Add them together >> '))
    answers = (user_answer, actual_answer)
    return answers

def minus():
    comp_num1 = random.randint(25,50)
    comp_num2 = random.randint(1,25)
    print(f'{comp_num1} - {comp_num2} = ')
    actual_answer = comp_num1 - comp_num2
    user_answer = int(input('Minus them together >> '))
    answers = (user_answer, actual_answer)
    return actual_answer, user_answer

def check_answer(actual_answer, user_answer):
    if actual_answer == user_answer:
        print('Correct')
    else:
        print('Incorrect, the answer is', actual_answer)

def main():
    print('1)Addition \n2) Subtraction ')
    selection = int(input('Enter 1 or 2 >> '))
    if selection == 1:
        user_answer, actual_answer = addition()
        check_answer(user_answer, actual_answer)
    elif selection == 2 :
        user_answer, actual_answer = minus()
        check_answer(user_answer, actual_answer)
    else:
        print('Incorrect selection')

main()
```

```python
# Create a program that will allow the user to easily manage a list of names.
# You should display a menu that will allow them to add a name to the list,
# change a name in the list, delete a name from the list or view all the names
# in the list. There should also be a menu option to allow the user to end the program.
# If they have made a select an option that is not relevant, then it should
# display a suitable message. After they have made a selection to either add a name,
# change a name, delete a name or view all the names, they should see the menu
# again without having to restart the program. The program should be made as easy to use as possible.
menu = []

def add():
    user = input('What menu you want to add >> ')
    menu.append(user)
    return menu

def change():
    num = 0 
    for i in menu:
        print(num, i)
    num += 1
    select_num = int(input('Enter the number of the name you want to change : '))
    name = input('Enter a new name >> ')
    menu[select_num] = name
    return menu

def delete():
    num = 0 
    for i in menu:
        print(num, i)
    num += 1
    select_num = int(input('Enter the number of the name you want to delete : '))   
    del menu[select_num]
    return menu

def view():
    num = 0 
    for i in menu:
        print(num, i)
    num += 1

def main():
    again = 'y'
    while again == 'y':
        print('1) Add \n2) Change \n3) Delete \n4) View')
        selection = int(input('Enter a number between 1 and 4 >> '))
        if selection == 1:
            names = add()
        elif selection == 2:
            names = change()
        elif selection == 3:
            names = delete()
        elif selection == 4:
            names = view()
        elif selection == 5:
            again = 'n'
        else:
            print('Incorrect selection')
```
