```python
# (영어) 알파벳과 숫자를 섞어서 입력 받아 숫자와 문자를
# 분리하여 출력하는 프로그램을 작성하시오

data = input('알파벳과 숫자를 섞어서 입력하시오 >> ')
al = ''
no = ''

for i in data:
    if i.isalpha() :
        al += i
    elif i.isdigit() :
        no += i

print(f'문자 : ' + al)
print(f'숫자 : ' + no)

```

```python
# 문자와 숫자의 갯수를 포함하여 출력하시오


data = input('알파벳과 숫자를 섞어서 입력하시오 >> ')
al = ''
no = ''

for i in data:
    if i.isalpha() :
        al += i
    elif i.isdigit() :
        no += i

print(f'문자 ({len(al)}) : {al}')
print(f'숫자 ({len(no)}) : {no}')
```

```python
# 문자와 숫자의 갯수를 포함하여 출력하시오
# 단, isalpha(), isdigit()

data = input('알파벳과 숫자를 섞어서 입력하시오 >> ')
al = ''
no = ''

for i in data:
    if i >= 'A' and i <= 'Z':
        al += i

    elif i >= 'a' and i <= 'z':
        al += i
        
    elif '1' <= i and i <= '9':
        no += i

print(f'문자 ({len(al)}) : {al}')
print(f'숫자 ({len(no)}) : {no}')
```

# 알고리즘
알고리즘은 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어나 생각

알고리즘 표현 방법은 수도코드(pseudo code), 플로우차트 (flow chart) 등이 있음

# 프로그램 작성 순서
기획 - 설계 (알고리즘) - 프로그래밍 - 디버그 (버그 잡는 것) - 문서작성
모든 프로그램은 버그가 필연적으로 일어나게 됨
전체 시간에 있어서 알고리즘 단계에서 설계를 대략 하게 되면 디버그 단계에서 문제가 많이 발생해 시간이 들 수밖에 없음
좋은 알고리즘을 구현해놓으면 추후 문제 여지가 많이 줄어 디버그에서의 시간 낭비가 줄어듦

# 좋은 알고리즘이란?
알기 쉽고, 속도가 빠르고, 효율적이고(메모리를 적게 사용하는 것), 재이용이 쉬움

# 왜 알고리즘을 써야하고 공부해야 하는가?
좋은 프로그램을 구현할 수 있고 좋은 프로그램인지 판단할 수 있음
프로그램 작성을 효율적으로 할 수 있음
프로그래밍 기술을 향상시키는데 큰 도움이 됨

# 알고리즘의 조건
결과를 얻을 수 있어야 하고 반드시 종료되어야 함

# 알고리즘 세가지 기본형
- 순차구조 (가장 많이 씀) : 처음부터 순서대로 처리
- 선택구조 (yes, no) : 조건식 if 로 표현
- 반복구조 (햄버거) : for문이나 while문으로 표현

# 순서도 (Flow Chart)
프로그래밍 언어를 사용하지 않고 알고리즘을 기술하는 방법
처리와 절차를 시각적으로 표현함
자신이 생각하고 있는 알고리즘이 올바른지, 처리와 절차에 문제가 없는지를 시각적으로 확인할 수 있음
다른 사람에게도 쉽게 전달 가능하다는 특징이 있음

# 순서도 주의 사항
처리 기호나 판단 기호가 많아지면 흐름이 복잡하게 꼬임

1. 입력은 반드시 위에서 들어가도록 만듦
2. 판단기호로부터 나오는 흐름선에는 Yes, no를 확실하게 기입함
3. 입력선이나 출력선이 없는 처리 기호는 만들지 않음

# 유명한 알고리즘이란?

유명한 알고리즘을 배우는 것은 프로그래밍 기술 향상에 큰 도움이 됨

1. 탐색 : Linear Search, Binary Search, Hash Search
2. 정렬 : Selection Sort, Bubble Sort, Insertion Sort, Quick Sort
3. Sieve of Eratosthenes, Euclid Algorithm

가우스 소거법, 사다리꼴 법칙, 디익스트라 알고리즘, 이분법, 뉴턴법
KMP 알고리즘, BM 알고리즘

# 이진 탐색법의 알고리즘
1. 가운데 요소를 선택하는 처리 (인덱스의 평균을 계산하여 구함)
2. 가운데 데이터와 찾는 데이터를 비교하는 처리
3. 탐색 범위를 절반으로 줄이는 처리

# 요소의 갯수가 짝수일 때는 어떻게 처리?
 0  1  2  3  4  5
11 13 17 19 23 29

- 요소의 갯수가 짝수일 때는 가운데 요소가 2개라는 것을 알 수 있음
- 이 경우 어느 쪽을 선택해야 한다고 정해진 것은 없음
- 단지 둘 중 하나만 선택하면 됨
- 따라서 같은 공식으로 계산해보면 (0 + 5) / 2 = 2.5
- 소수를 정수화하는 방법으로는
  1. 반올림하는 방법
  2. 올림하는 방법
  3. 버림하는 방법
  제일 간단하고 쉬운 처리 방법 : 3. 버림 < 정수형 변수에 넣기만 하면 소수 부분은 잘려 나가기 때문에
                                         쉽게 정수 부분만 얻을 수 있음

        

# 해시 탐색법

데이터의 '내용'과 저장한 곳의 '요소'를 연계해두어 극히 짧은 시간 안에 검색할 수 있도록 고안된 알고리즘

해시 값 = 데이터 % 저장 공간의 수 <- 해시함수
미리 해시 함수를 사용하여 데이터를 저장하는 장소를 정해서 검색 시간을 놀라울 정도로 단축시킬 수 있음

# 해시 탐색 알고리즘
1. 해시함수를 통해 데이터를 저장하는 알고리즘
2. 해시함수를 통해 데이터를 검색하는 알고리즘

배열을 2개 준비
정렬 전에 준비된 데이터의 배열과 데이터를 저장할 배열을 준비
데이터를 저장할 배열의 수는 데이터 갯수의 1.5 배에서 2배로 준비하는 것이 좋음
검색 속도와 메모리 효율 면에서 적당 크기는 1.5배에서 2배로 알려져 있음

# 단순 선택법
1. 탐색 범위의 최소값을 찾는 처리
2. 찾을 최소값을 맨 앞의 요소와 교환하는 처리

# 단순 교환법
1. 인접한 2개의 데이터를 교환하는 처리를 반복하여 전체를 정렬
2. 속도가 느린 알고리즘

# 단순 교환법의 알고리즘
1. 오른쪽 끝 요소부터 차례로 인접한 데이터를 오름차순으로 정렬하여 교환함
2. 왼쪽 끝부터 순서대로 하나씩 데이터를 정렬하여 확정시킴
